```python
import pandas as pd
import requests
import numpy as np
import plotly.express as px
pd.options.plotting.backend = "plotly"
```

```python
# Define query URL
get_tsunami_data_URL = "http://127.0.0.1:5000/get_tsunami_data"

# Get the tsunami data via GET
response = requests.get(get_tsunami_data_URL).json()

# Create dataframe from response list
tsunami_df = pd.DataFrame(response)
```

```python
# Filter for only definite tsunamis
# - Tsunami Event Validity = 4, meaning Definite Tsunami
# - Doubtful Runup = "n",       meaning runup entry was not doubtful
tsunami_df = tsunami_df.loc[((tsunami_df["Tsunami Event Validity"] == "4") & (tsunami_df["Doubtful Runup"] == "n")),:]

# Drop unneeded columns
tsunami_df.drop(['Tsunami Event Validity', 'Tsu Src', 'Vol', 'More Info', 'Doubtful Runup', 'Period', 'First Motion', 'Search Parameters'], axis=1, inplace=True)

# Replace blanks with zeroes
tsunami_df = tsunami_df.replace(r'^\s*$', 0, regex=True)
```

```python
# Define a conversion dictionary
conversion_dict = {"Damage $Mil" : "float64",            
                   "Damage Description" : "int64",
                   "Death Description" : "int64",
                   "Deaths" : "int64",
                   "Distance From Source (km)" : "float64",
                   "Dy" : "int64",
                   "Earthquake Magnitude" : "float64",
                   "Houses Damaged" : "int64",
                   "Houses Damaged Description" : "int64",
                   "Houses Destroyed" : "int64",
                   "Houses Destroyed Description" : "int64",
                   "Hr" : "int64",
                   "Initial Wave Arrival Day" : "int64",
                   "Initial Wave Arrival Hour" : "int64",
                   "Initial Wave Arrival Minute" : "int64",
                   "Injuries" : "int64",
                   "Injuries Description" : "int64",
                   "Latitude" : "float64",
                   "Longitude" : "float64",
                   "Max Inundation Distance (m)" : "float64",
                   "Max Wave Arrival Day" : "int64",
                   "Max Wave Arrival Hour" : "int64",
                   "Max Wave Arrival Minute" : "int64",
                   "Maximum Water Height (m)" : "float64",
                   "Measurement Type" : "int64",
                   "Missing" : "int64",
                   "Missing Description" : "int64",
                   "Mn" : "int64",
                   "Mo" : "int64",
                    "Sec" : "float64",
                   "Travel Hours" : "int64",
                   "Travel Minutes" : "int64",
                   "Tsunami Cause Code" : "int64",
                   "Year" : "int64"
                  }

# Convert column datatypes, ignore errors and keep any NaNs if they exist
tsunami_df = tsunami_df.astype(conversion_dict)
```

```python
# Rearrange columns
column_order = ["Year",
                "Mo",
                "Dy",
                "Hr",
                "Mn",
                "Sec",
                "Tsunami Cause Code",
                "Earthquake Magnitude",
                "Country",
                "Area",
                "Location Name",
                "Latitude",
                "Longitude",
                "Distance From Source (km)",
                "Initial Wave Arrival Day",
                "Initial Wave Arrival Hour",
                "Initial Wave Arrival Minute",
                "Travel Hours",
                "Travel Minutes",
                "Max Wave Arrival Day",
                "Max Wave Arrival Hour",
                "Max Wave Arrival Minute",
                "Maximum Water Height (m)",
                "Max Inundation Distance (m)",
                "Measurement Type",
                "Deaths",
                "Death Description",
                "Missing",
                "Missing Description",
                "Injuries",
                "Injuries Description",
                "Damage $Mil",
                "Damage Description",
                "Houses Destroyed",
                "Houses Destroyed Description",
                "Houses Damaged",
                "Houses Damaged Description"]

tsunami_df = tsunami_df.reindex(columns=column_order)
```

#### Histogram: Number of tsunamis by year

```python
# Make copy of dataframe
plot_df = tsunami_df.copy()

# Filter for only true tsnuamis
plot_df = plot_df.loc[(plot_df["Maximum Water Height (m)"]> 0), :]

fig = px.histogram(plot_df, x="Year", nbins=20, title="Histogram: Number of Tsunamis by Year")
fig.update_xaxes(nticks=20)
fig.show()
```

#### Histogram: Number of tsunamis by month

```python
# Make copy of dataframe
plot_df = tsunami_df.copy()

# Filter for only true tsnuamis
plot_df = plot_df.loc[(plot_df["Maximum Water Height (m)"]> 0), :]

fig = px.histogram(plot_df, x="Mo", color="Year", title="Histogram: Number of Tsunamis by Month")
fig.update_layout(
    xaxis = dict(tickmode = 'linear',
                tick0 = 1,
                dtick = 1))
fig.show()
```

#### Histogram: Number of tsunamis by cause code

```python
# Make copy of dataframe
plot_df = tsunami_df.copy()

# Filter for only true tsnuamis
plot_df = plot_df.loc[(plot_df["Maximum Water Height (m)"]> 0), :]

fig = px.histogram(plot_df, x="Tsunami Cause Code", title="Histogram: Number of Tsunamis by Tsunami Cause Code")
fig.show()

print("Tsunami Cause Codes:")
print("0	Unknown")
print("1	Earthquake")
print("2	Questionable Earthquake")
print("3	Earthquake and Landslide")
print("4	Volcano and Earthquake")
print("5	Volcano, Earthquake, and Landslide")
print("6	Volcano")
print("7	Volcano and Landslide")
print("8	Landslide")
print("9	Meteorological")
print("10	Explosion")
print("11	Astronomical Tide")
```

#### Histogram: Number of tsunamis by country

```python
# Make copy of dataframe
plot_df = tsunami_df.copy()

# Filter for only true tsnuamis
plot_df = plot_df.loc[(plot_df["Maximum Water Height (m)"]> 0), :]

fig = px.histogram(plot_df, y="Country", title="Histogram: Number of Tsunamis by Country", height=1500).update_yaxes(categoryorder="total ascending")
fig.show()
```

#### Scatter: Earthquake Magnitude vs. Maximum Water Height (m)

```python
# Build dataframe of scatter data
x_axis = tsunami_df.loc[tsunami_df["Earthquake Magnitude"] > 0 , "Earthquake Magnitude"]
y_axis = tsunami_df.loc[tsunami_df["Earthquake Magnitude"] > 0 , "Maximum Water Height (m)"]

plot_dict = {"Earthquake Magnitude" : x_axis,
             "Maximum Water Height (m)" : y_axis}

plot_df = pd.DataFrame(plot_dict)

fig = plot_df.plot.scatter(x="Earthquake Magnitude", y="Maximum Water Height (m)", title="Earthquake Magnitude vs. Maximum Water Height (m)")
fig.show()
```
